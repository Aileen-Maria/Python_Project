```python
import pandas as pd
df = pd.read_csv(r"C:\Users\zerxh\Downloads\myexcel - myexcel.csv.csv")
df.head()

```

```python
df['height'] = np.random.randint(150, 181, df.shape[0])

df.head()
```

```python
team_distribution = df["Team"].value_counts()
team_percentage = (team_distribution / len(df)) * 100

print("Employee Distribution Across Teams:\n", team_distribution)
print("\nPercentage Distribution:\n", team_percentage)
```

```python
position_distribution = df["Position"].value_counts()

print("Employee Distribution Across Positions:\n", position_distribution)
```

```python
import matplotlib.pyplot as plt

bins = [18, 25, 35, 45, 55, 65]
labels = ["18-25", "26-35", "36-45", "46-55", "56-65"]

df["age_group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)

age_group_distribution = df["age_group"].value_counts()

print("Predominant Age Group:\n", age_group_distribution)
```

```python
team_salary = df.groupby("Team")["Salary"].sum()

position_salary = df.groupby("Position")["Salary"].sum()

print("Team with Highest Salary Expenditure:\n", team_salary.idxmax(), "-", team_salary.max())
print("Position with Highest Salary Expenditure:\n", position_salary.idxmax(), "-", position_salary.max())
```

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary Correlation")
plt.show()
```

```python
team_distribution.plot(kind="bar", title="Employee Distribution Across Teams", ylabel="Count", xlabel="Team", colormap="viridis")
plt.show()

```

```python
position_distribution.plot(kind="pie", autopct='%1.1f%%', title="Employee Distribution Across Positions", startangle=90, colormap="coolwarm")
plt.ylabel("")
plt.show()
```

```python
team_salary.sort_values().plot(kind="bar", title="Salary Expenditure by Team", ylabel="Total Salary", xlabel="Team", colormap="magma")
plt.show()
```

```python
position_salary.sort_values().plot(kind="bar", title="Salary Expenditure by Position", ylabel="Total Salary", xlabel="Position", colormap="cividis")
plt.show()
```

```python
print("Insights from the Analysis:")
print("- The team with the highest number of employees is:", team_distribution.idxmax())
print("- The most common position in the company is:", position_distribution.idxmax())
print("- The predominant age group is:", age_group_distribution.idxmax())
print("- The team with the highest salary expenditure is:", team_salary.idxmax())
print("- The position with the highest salary expenditure is:", position_salary.idxmax())
print("- From the Age vs Salary scatter plot, we can observe if salary increases with age.")
```
